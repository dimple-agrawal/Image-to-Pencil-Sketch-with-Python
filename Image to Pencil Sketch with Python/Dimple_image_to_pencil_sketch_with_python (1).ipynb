## ABOUT THE PROJECT 

**IMAGE TO PENCIL SKETCH WITH PYTHON** is a really fun beginner level task any python enthusiast can do using only the ***OpenCV *** library in python.
It's a four step process step that will convert your uploaded RGB image into a pencil sketch image.

The following are the steps to be performed:-


1.  Convert the RGB color image to grayscale.
2.  Invert the grayscale image to get negative.
3.  Apply Gaussian Blur to the negative from step 2.
4.  Invert the blurred image.








*#Import the cv2 library*

# IMPORT REQUIRED MODULE

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

*#Download and upload the desired image in the file menu.*
Read and show the image


```python
img = cv2.imread("flag.jpg")
```

```python
cv2_imshow(img)
```

*#Since the image size is too big, we'll resize the image to half its original size*

```python
scale_percent=0.50
width=int(img.shape[1]*scale_percent)
height=int(img.shape[0]*scale_percent)
dimension=(width,height)
```

```python
resized=cv2.resize(img,dimension,interpolation=cv2.INTER_AREA)
```

```python
cv2_imshow(resized)
```

**#Step 1**
Convert the image to gray scale

```python
gray_img = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
```

**#Step 2**
*Invert the gray image* 

```python
inverted_gray_img=cv2.bitwise_not(gray_img)
```

```python
cv2_imshow(inverted_gray_img)
```

**#Step 3**
Blur the image using Gaussian function

What is Gaussian image?


> In image processing, a **Gaussian blur** (also known as Gaussian smoothing) is the result of blurring an image by a Gaussian function (named after mathematician and scientist Carl Friedrich Gauss).



It is a widely used effect in graphics software, typically to reduce image noise and reduce detail. 

```python
blurred_img=cv2.GaussianBlur(inverted_gray_img,(21,21),sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(blurred_img)
```

**#Step 4**
*Invert the blurred image*

Blend the gray-scale image from step 1 with the blurred negative from step 3 using a color dodge.

```python
def dodgev2(x,y):
  return cv2.divide(x, 255-y, scale=256)
```

```python
final_img= dodgev2(gray_img, blurred_img)
```

***Here's your final image***

```python
cv2_imshow(final_img)
```
